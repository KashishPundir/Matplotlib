TYPES OF MACHINE LEARNING:
1) Supervised Machine Learning (SVM): Supervised machine learning uses labeled datasets to train algorithms to classify data or predict outcomes.
Common algorithms: Linear Regression, Logistic Regression, Support Vector Machines (SVM), Decision Trees, Random Forests, and Neural Networks.
Example applications: Spam detection, image classification, and medical diagnosis.

2) Unsupervised Learning: In unsupervised learning, the algorithm is provided with unlabeled data and must find patterns and relationships within the data.
Common algorithms: K-Means Clustering, Hierarchical Clustering, Principal Component Analysis (PCA), and Association Rules.
Example applications: Customer segmentation, anomaly detection, and market basket analysis.

3) Semi-Supervised Learning: Semi-supervised learning falls between supervised and unsupervised learning. 
It uses a small amount of labeled data and a large amount of unlabeled data.
Common algorithms: Semi-Supervised Support Vector Machines, Label Propagation.
Example applications: Web page classification, and text document classification.

4) Reinforcement Machine Learning: Reinforcement machine learning is a type of machine learning model that is similar 
to supervised learning but does not use sample data to train the algorithm. This model learns by trial and error.
Common algorithms: Q-Learning, Deep Q-Networks (DQN), Policy Gradient Methods.
Example applications: Game playing (e.g., AlphaGo), robotics, and autonomous vehicles.

Regression models are of following two types

Multiple regression model  As name implies, in this regression model the predictions are formed from multiple features of the data.

```python
# creates linear regression plot
import numpy as np
from sklearn import linear_model
import sklearn.metrics as sm
import matplotlib.pyplot as plt
input = r'C:\linear.txt'
input_data = np.loadtxt(input, delimiter=',')
X, y = input_data[:, :-1], input_data[:, -1]
training_samples = int(0.6 * len(X))
testing_samples = len(X) - training_samples
X_train, y_train = X[:training_samples], y[:training_samples]
X_test, y_test = X[training_samples:], y[training_samples:]
reg_linear = linear_model.LinearRegression()
reg_linear.fit(X_train, y_train)
y_test_pred = reg_linear.predict(X_test)
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, y_test_pred, color = 'black', linewidth = 2)
```

# LINEAR REGRESSION :
Linear regression may be defined as the statistical model that analyzes the linear relationship between a dependent variable with given set of independent variables.
The relationship can be represented with the help of following equation  Y=mX+b
Here,
Y is the dependent variable we are trying to predict
X is the dependent variable we are using to make predictions
m is the slop of the regression line which represents the effect X has on Y.
b is a constant, known as the Y-intercept. If X = 0, Y would be equal to b.

*Positive Linear Relationship*
A linear relationship will be called positive if both independent and dependent variable increases.
*Negative Linear Relationship*
A linear relationship will be called positive if independent increases and dependent variable decreases.

Multiple Linear Regression models always includes the errors in the data known as residual error which changes the calculation as follows 

h(xi)=b0+b1xi1+b2xi2++bpxip+ei
*We can also write the above equation as follows *
yi=h(xi)+eiorei=yih(xi)

# Mean Squared Error (MSE)
Mean Squared Error (MSE) is a measure of the average of the squares of the errorsthat is, the average squared difference between the estimated values (^) and the actual value (y).
Formula:
MSE=1/(=1 to (^))^2
where:i are the actual values
      ^ are the predicted values
      n is the number of observations
The MSE gives us an idea of how close the predicted values are to the actual values. A lower MSE indicates that the model's predictions are closer to the actual values.

# Coefficient of Determination (R):
The coefficient of determination, commonly known as R, is a statistical measure that explains the proportion of the variance in the dependent variable that is predictable from the independent variable(s).
Formula:
2=1(=1 to (^)^2)/(=1 to ()^2)
where: are the actual values
       cap  are the predicted values
       is the mean of the actual values

---> ^2 ranges from 0 to 1.
An ^2 of 1 indicates that the regression predictions perfectly fit the data.
An ^2 of 0 indicates that the model does not explain any of the variability in the response data around its mean.
An ^2 between 0 and 1 indicates the proportion of the variance in the dependent variable that is predictable from the independent variable(s).

# 2) K-Nearest Neighbors (KNN) : 
KNN is a supervised learning algorithm that can be used for both classification and regression problems. 
The main idea behind KNN is to find the k-nearest data points to a given test data point and use these nearest neighbors to make a prediction.
k=number of neighbours to consider for classification of test point 
---> The distance metric used to measure the similarity between two data points is an essential factor that affects the KNN algorithm's performance. The most commonly used distance metrics are Euclidean distance, Manhattan distance, and Minkowski distance. 
EX :  To classify a new flower using KNN:
1) Choose k = 3.
2) Calculate the distance between the new flower and all flowers in the training set.
3) Identify the 3 closest flowers.
4) Assign the species that is most common among these 3 neighbors to the new flower.

----> Minkowski Distance: 
(=1 to  ^)^1/

# 5) Support Vector Machine (SVM) :
The goal of SVM is to find a hyperplane that separates the data points into different classes. A hyperplane is a line in 2D space, a plane in 3D space, or a higher-dimensional surface in n-dimensional space. The hyperplane is chosen in such a way that it maximizes the margin, which is the distance between the hyperplane and the closest data points of each class. The closest data points are called the support vectors.
The distance between the hyperplane and a data point "x" can be calculated using the formula 
distance = (w . x + b) / ||w|| 
where "w" is the weight vector, "b" is the bias term, and "||w||" is the Euclidean norm of the weight vector. The weight vector "w" is perpendicular to the hyperplane and determines its orientation, while the bias term "b" determines its position

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# load the iris dataset
iris = load_iris()

# split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris.data,
iris.target, test_size=0.2, random_state=42)

# create an SVM classifier with a linear kernel
svm = SVC(kernel='linear')

# train the SVM classifier on the training set
svm.fit(X_train, y_train)

# make predictions on the testing set
y_pred = svm.predict(X_test)

# calculate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

# 6) Random Forest :
Random Forest is a machine learning algorithm that uses an ensemble of decision trees to make predictions.The algorithm was first introduced by Leo Breiman in 2001.

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
# Loading the iris dataset
iris =sns.load_dataset("iris")

# Separating the features and target variable
X = iris.iloc[:, :-1]
y = iris.iloc[:, -1]

# Splitting the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size=0.35, random_state=42)

# Creating the Random Forest classifier object
rfc = RandomForestClassifier(n_estimators=100)

# Training the model on the training data
rfc.fit(X_train, y_train)
# Making predictions on the test data
y_pred = rfc.predict(X_test)
# Importing the metrics library
from sklearn.metrics import accuracy_score, precision_score,recall_score, f1_score

# Calculating the accuracy, precision, recall, and F1-score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```

# 7) Confusion Matrix :
 A confusion matrix is nothing but a table with two dimensions viz. "Actual" and "Predicted" and furthermore, both the dimensions have "True Positives (TP)", "True Negatives (TN)", "False Positives (FP)", "False Negatives (FN)" as shown below 
                                  Actual 
                     1                           0

            1    True Positives(TP)      False Positives(FP)
Predicted  

            0    False Negatives(FN)     True Negatives(TN)

Confusion Matrix will tell u what your machine laerning algorithm did right and what it did wrong.

```python
from sklearn.metrics import confusion_matrix

# Actual values
y_actual = [0, 1, 0, 1, 1, 0, 0, 1, 1, 1]

# Predicted values
y_pred = [0, 1, 0, 1, 0, 1, 0, 0, 1, 1]

# Confusion matrix
cm = confusion_matrix(y_actual, y_pred)
print(cm)
```

```python
import seaborn as sns

# Plot confusion matrix as heatmap
sns.heatmap(cm, annot=True, cmap='summer')
```
