```python
A=np.array([[1,2],
            [3,4]])
# create a vector
x=np.array([1,1])
b=np.dot(A,x)  # 3,7
plt.plot([0,x[0]],[0,x[1]],label="Original Vector",color="blue",marker="o")  # (0,1),(0,1) ---> (0,0),(1,1)
plt.plot([0,b[0]],[0,b[1]],label="Transformed Vector",color="red",marker="o") # (0,3),(0,7)---> (0,0),(3,7)
plt.xlim(-1,5)
plt.ylim(-1,8)
plt.title("Linear Transformation")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

# Optimizer in Scipy :
Optimizers are a set of procedures defined in SciPy that either find the minimum value of a function, or the root of an equation.

```python
# This line of code imports a specific function, `root`, from the `scipy.optimize` module. 
from scipy.optimize import root
from math import cos
def eqn(x):
  return x + cos(x)
myroot = root(eqn, 0)
print(myroot.x,"\n")
print(myroot)
```

```python
from scipy import stats    #  stats --> module
import numpy as np
import matplotlib.pyplot as plt
data=np.random.normal(loc=0,scale=1,size=1000)
mean_value=np.mean(data)
median_value=np.median(data)
mode_value=float(stats.mode(data)[0])
plt.hist(data,bins=30,alpha=0.5,color="b",label="Data")
plt.axvline(mean_value,color="r",linestyle="--",label="Mean")
plt.axvline(median_value,color="g",linestyle="--",label="Median")
plt.axvline(mode_value,color="y",linestyle="--",label="Mode")
plt.legend()
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Central Tendencies Visualization")
plt.show()

```

```python
# Pararmeters for normal Distribution
mu=0
sigma=1
sample_size=1000
data=np.random.normal(mu,sigma,sample_size)
plt.figure(figsize=(8,6))
plt.hist(data,bins=30,density=True,alpha=0.6,color="blue",edgecolor="black")
plt.title("Histogram of random samples for normal distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

xmin,xmax=plt.xlim()
x=np.linspace(xmin,xmax,100)
p=1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-mu)**2/(2*sigma*2))
plt.plot(x,p,"k",linewidth=2)
plt.show()
```

```python
from scipy.stats import norm
mu=0
sigma=1     
plt.hist(data,bins=30,density=True,alpha=0.6,color="blue",edgecolor="black")
xmin,xmax=plt.xlim()
x=np.linspace(xmin,xmax,100)
p=norm.pdf(x,mu,sigma)    # pdf --> probablity density function
plt.plot(x,p,"k",linewidth=2)
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram and PDF of a normal distribution")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
A=np.array([[1,2],
            [3,4]])
# create a vector
x=np.array([1,1])
b=np.dot(A,x)  # 3,7
plt.quiver(0,0,x[0],x[1],angles="xy",scale_units="xy",scale=1,label="Original Vector",color="blue")  # (0,1),(0,1) ---> (0,0),(1,1)
plt.quiver(0,0,b[0],b[1],angles="xy",scale_units="xy",scale=1,label="Resultant Vector",color="red") # (0,3),(0,7)---> (0,0),(3,7)
plt.xlim(0,b[0]+1)
plt.ylim(0,b[1]+1)
plt.title("Result of Matrix over mutiplication")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.legend()
plt.show()
```
